## 1. Identify the components

### 1.1 what are the input?
- `quiz_1` - mark for Quiz 1 (entered by the user)
- `quiz_2` - mark for Quiz 2 (entered by the user)
- `quiz_3` - mark for Quiz 3 (entered by the user)
- `choice` - Y/N response that decides whether to repeat the program for another student

### 1.2 What is the process?
1. Convert each entered mark from text into a number (float)
2. Add the three marks together and divide by 3 to calculate the average
3. Compare the average to the pass mark (50) to decide pass or fail
4. Ask the user whether they want to repeat the process for another student

### 1.3 What is the output?
- The calculated **average** mark, displayed on screen
- A **PASS** or **Fail** result, displayed on screen 
- A final "program Ended" message once the user chooses to stop

## 2. Design The Algorithm

### 2.1 pseudocode

```
START
SET choice = "y" 
WHILE choice equals "y" Do
    INPUT quiz_1
    INPUT quiz_2
    INPUT quiz_3
    CALCULATE average = (quiz_1 + quiz_2 + quiz_3) / 3
    DISPLAY average
    IF average >= 50 THEN
        DISPLAY "PASS"
    ELSE
        DISPLAY "FAIL"
    ENDIF
    INPUT choice
ENDWHILE
DISPLAY "Program Ended"
END
```
    

### 2.2 Identify the algorithm structure

| Structure | Where it appears in the code | Why |
|---|---|---|
| **Sequence** | Reading `quiz_1`, quiz_2, quiz_3, then calculating `average` | Steps run one after another, in a fixed order |
| **Selection** | The `if average >= 50: ... else: ...` block | The program chooses between two different paths (PASS or FAIL) depending on a condition |
| **Iteration** | The `while choice == "y":` loop | The same block of code repeats for as many students as the user wants to enter |

## 3. The Python Program
The completed program is saved separately as `Student_Ave_Calc.py`. A copy is shown below for reference.

````python
choice = "y"

# This while loop keeps the program running as long as the user enters "y",
# allowing marks for multiple students to be entered one after another (Iteration)
while choice == "y":

    # Ask the user to enter each of the three quiz marks, converting the
    # text typed in by the user into a decimal number (float) so maths can be done on it
    quiz_1 = float ( input ( "Enter Quiz 1 mark: " ) )
    quiz_2 = float ( input ( "Enter Quiz 2 mark: " ) )
    quiz_3 = float ( input ( "Enter Quiz 3 mark: " ) )

    # Add the three marks together and divide by 3 to get the average,
    # then display the result to the user
    average = ( quiz_1 + quiz_2 + quiz_3 ) / 3
    print ( "Average mark:", average )

    # Check whether the average is a pass (50 or above) or a fail (Selection)
    if average >= 50:
        print ( "Result: PASS" )
    else:
        print ( "Result: FAIL" )

    # Ask the user if they want to enter marks for another student.
    # If they type anything other than "y", the while loop will stop
    choice = input ( "Continue? Select Y/N: " )

print ( "Program Ended" )
````
